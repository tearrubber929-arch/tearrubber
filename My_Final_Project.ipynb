Reading the CSV file

```python

# Step 1 : import library
import pandas as pd


# Step 2 : import data
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
cancer
```

Details

```python
cancer.info()
```

```python
cancer.describe()
```

Feature selection

```python
x=cancer.drop(['diagnosis','Unnamed: 32'],axis=1)
y=cancer['diagnosis']
```

```python
import matplotlib.pyplot as plt
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

# Initialize and fit the scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

# Create DataFrame with scaled values
df_standardized = pd.DataFrame(X_scaled, columns=x.columns, index=x.index)
```

```python
df_standardized
```

```python
df_standardized.describe()
```

```python
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score,
    roc_curve, auc
)
```

Training, Testing and Prediction

```python

#train test split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, train_size=0.7, random_state=2529
)
```

```python
#select model
model = LogisticRegression(max_iter=5000)
```

```python
#train or fit model
model.fit(X_train, y_train)
```

```python
#predict on test data
y_pred = model.predict(X_test)
```

Model Report

```python
# model evaluation
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
```

```python
# Step 9 : Feature Importance (Logistic Regression Coefficients)
import numpy as np

# Get absolute values of coefficients
feature_importance = pd.Series(
    np.abs(model.coef_[0]),
    index=x.columns
).sort_values(ascending=False)

# 📊 Plot Feature Importance
plt.figure(figsize=(10,6))
sns.barplot(x=feature_importance.values[:15],
            y=feature_importance.index[:15],
            palette="viridis")
plt.title("Top 15 Most Important Features in Cancer Prediction")
plt.xlabel("Importance (absolute coefficient value)")
plt.ylabel("Feature")
plt.show()
```

Data Visualization

```python
# 📊 Visualization 1: Confusion Matrix Heatmap
plt.figure(figsize=(6,4))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Benign','Malignant'],
            yticklabels=['Benign','Malignant'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()
```
